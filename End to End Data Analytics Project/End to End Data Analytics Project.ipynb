```python
order_df = pd.read_csv("./orders.csv",na_values=['Not Available', 'unknown'])
```

```python
order_df.columns = order_df.columns.str.lower()
```

```python
order_df.columns = order_df.columns.str.replace(' ',"_")
```

```python
order_df.head(5)
```

```python
order_df['discount'] = order_df['list_price'] * order_df['discount_percent']*.01 # -> for discount column
order_df['sale_price'] = order_df['list_price'] - order_df['discount'] # -> for sale_price column
order_df['profit'] = order_df['sale_price'] - order_df['cost_price'] # -> for discount column
order_df.head(5)
```

```python
order_df["order_date"] = pd.to_datetime(order_df['order_date'],format="%Y-%m-%d")
```

```python
order_df.drop(columns=['list_price','cost_price', 'discount_percent'], inplace=True) # inplace=true for del col permanently
```
